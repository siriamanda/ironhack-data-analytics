```python
# 2. How could we find out gender when this field is null? 

data['GENDER'].value_counts()
```

```python
# Let's consider the gender column.

# Can we guess in somehow what is the missing gender? NO.
# Can this column have any possible value compatible with a missing value? NO
# Therefore, we are forced to drop the corresponding rows.
```

```python
to_drop = data[~data['GENDER'].isin(['F','M'])].index.tolist()
#data = data[data['GENDER'].isin(['F','M'])]
data.drop(to_drop, inplace = True)
data.reset_index(drop=True)
```

```python
data['GENDER'].value_counts()
```

```python
# 3. The homeowner column has 2 values (H, U). Fill the null values accordingly

data['HOMEOWNR'].value_counts()
```

```python
np.unique(data['HOMEOWNR']).tolist()
```

```python
data['HOMEOWNR'] = np.where(data['HOMEOWNR'] == ' ','U',data['HOMEOWNR'])
```

```python
data['MDMAUD'].value_counts()
```

```python
mdmaud_values = list(np.unique(data['MDMAUD']))
mdmaud_values
```

```python
mdmaud_values.remove('XXXX')
```

```python
def clean_MDMAUD(x):
    if x in mdmaud_values:
        return 'other'
    else:
        return x

data['MDMAUD'] = list(map(clean_MDMAUD, data['MDMAUD']))
```

```python
# Simpler versions removing the list-step:

def grouping(x):
    if x == "XXXX":
        return "XXXX"
    else:
        return "other"

data["MDMAUD"] = list(map(grouping, data["MDMAUD"]))


def clean_mdmaud(x):
    if x != 'XXXX':
        return 'other'
    else:
        return x

data['MDMAUD'] = list(map(clean_mdmaud, data['MDMAUD']))

# More efficient

data['MDMAUD'] = np.where(data['MDMAUD'] != 'XXXX', 'other', 'XXXX')
```
