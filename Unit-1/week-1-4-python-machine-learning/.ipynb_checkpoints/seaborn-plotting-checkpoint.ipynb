```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# print()
```

```python
# Scatterplot
# Looking at the correlation of the amount of the bill and the amount of the tip

plt.scatter(x = tips['total_bill'], y = tips['tip'])
```

```python
# hue is a block color - can be one or the other. Changes the color of the spcified variable. 
# Use an outerboard when making visualisations that overlay 

sns.scatterplot(data = tips, x = 'total_bill', y = 'tip', hue = 'time')
```

```python
# style changes the shape of of the specified variable

sns.scatterplot(data = tips, x = 'total_bill', y = 'tip', hue = 'time', style = 'time')
```

```python
sns.scatterplot(data = tips, x = 'total_bill', y = 'tip', hue = 'day', style = 'time')
```

```python
sns.scatterplot(data = tips, x = 'total_bill', y = 'tip', hue = 'size')
```

```python
# Changing the color palette

sns.scatterplot(data = tips, x = 'total_bill', y = 'tip', hue = 'size', palette = 'BrBG')
```

```python
sns.scatterplot(data = tips, x = 'total_bill', y = 'tip', hue = 'size', size = 'size')
    
```

```python
# legend = 'full' enforce all data to be shown

sns.scatterplot(data = tips, x = 'total_bill', y = 'tip', hue = 'size', size = 'size',
               sizes = (20, 200), legend = 'full')
    
```

```python
# Relational plot, relplot allows comparing plots next to each other

sns.relplot(data = tips, x = 'total_bill', y = 'tip', col = 'time', hue = 'day', style = 'day', kind = 'scatter')
```
