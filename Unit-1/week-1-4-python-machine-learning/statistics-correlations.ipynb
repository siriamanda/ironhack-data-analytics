```python
# Need to clean data before this stage

x = data['ic2']
y = data['ic3']
stats.pearsonr(x, y)       # Gives the correlation coefficioent and the p-value
```

```python
# Correlations matrix
# Take all of the columns and look for correlations

corr_data = data.drop(['id', 'state', 'gender'], axis = 1)
correlations_matrix = corr_data.corr()

correlations_matrix

# Check that thematrix is working - the value is '1' diagonally 
```

```python
sns.heatmap(correlations_matrix, annot = True)
plt.show()
```

```python
mask = np.zeros_like(correlations_matrix)
mask[np.triu_indices_from(mask)] = True
fig, ax = plt.subplots(figsize=(10, 8))
ax = sns.heatmap(correlations_matrix, mask=mask, annot=True)
plt.show()
```

```python
# Create a crosstab table

crosstab = pd.crosstab(data['state'], data['gender']) 

crosstab
```

```python
# Chi-square test

[statistic, p_value, dof, arr] = stats.chi2_contingency(crosstab)

print(statistic)
print(p_value)
print(dof)

# first value is the ch-sq test statistics, the second value is the p-value
# third value is the degrees of freedom and 4th value an array that contains the expected cell count
```
