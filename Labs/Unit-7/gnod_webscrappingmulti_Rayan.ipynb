```python
# 1. import libraries
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# 2. url: we start with the 'second' page
url = "https://musicbrainz.org/series/b3484a66-a4de-444d-93d3-c99a73656905?page=1"
```

```python
# 3. download html with a get request
response = requests.get(url)
response.status_code # 200 status code means OK!
```

```python
# 4.1. parse html (create the 'soup')
soup = BeautifulSoup(response.content, "html.parser")
# 4.2. check that the html code looks like it should
soup
```

```python
from random import randint
import time

#Assembling the script to send and store multiple requests
iterations = range(1,6)

songs = []
artists = []

for i in iterations:
    # assemble the url:
    start_at= str(i)
    url = "https://musicbrainz.org/series/b3484a66-a4de-444d-93d3-c99a73656905?page=" + start_at

    # download html with a get request:
    response = requests.get(url)

    # monitor the process by printing the status code
    print("Status code: " + str(response.status_code))

    # store response into "pages" list
    songs.append(soup.select("td:nth-child(2) > a > bdi"))
    artists.append(soup.select("td:nth-child(3) > a > bdi"))

    # respectful nap:
    wait_time = randint(1,4)
    print("I will sleep for " + str(wait_time) + " second/s.")
    time.sleep(wait_time)
```
