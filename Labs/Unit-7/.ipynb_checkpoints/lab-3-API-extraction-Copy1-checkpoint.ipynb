```python
# Import the libraries

import spotipy
from spotipy.oauth2 import SpotifyClientCredentials

# some extra useful libraries

import pandas as pd
import json
import pprint
import getpass
```

```python
# Initialize SpotiPy with user credentias using getpass

client_id = getpass.getpass(prompt="Spotify client_id: ")
client_secret = getpass.getpass(prompt="Spotify client_secret: ")

sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(
    client_id=client_id,
    client_secret=client_secret)
                    )
```

```python
# Testing to retrieve playlist

playlist = sp.playlist_tracks("https://open.spotify.com/playlist/37i9dQZF1E4p5ofi3OATFS")
```

```python
# Define a function to extract a playlist and return the uri

def track_ids_from_playlist(playlist_id):       # We give the playlist_id as argument
    
    final_results = []
    result = sp.playlist_tracks(playlist_id)
    final_results.extend(result["items"])
    
    while result["next"]:
        
        result = sp.next(result)
        final_results.extend(result["items"])
        
    ids = [track["track"]["id"] for track in final_results]
    
    return [i for i in ids if i]
```

```python
track_ids_from_playlist("37i9dQZF1DX3Ebqev5IkYU")
```

```python
# Define a function to iterate through the list of track ids and return a dataframe with the audio features

def audio_features_df(playlist_id):
    track_ids = track_ids_from_playlist(playlist_id)
    aud_feat = [sp.audio_features(track)[0] for track in track_ids]      # iterate through the list of ids and retrieve the audio features
    return pd.DataFrame(aud_feat)
```

```python
a_features = audio_features_df("https://open.spotify.com/playlist/37i9dQZF1E4p5ofi3OATFS")
a_features.head()
```

```python
# Create a large dataframe

playlists = ["https://open.spotify.com/playlist/37i9dQZF1E4p5ofi3OATFS",
             "https://open.spotify.com/playlist/37i9dQZF1DWVWiyE9VDkCO",
             "https://open.spotify.com/playlist/37i9dQZF1DX4uPi2roRUwU",
             "https://open.spotify.com/playlist/37i9dQZF1DWSJHnPb1f0X3",
             "https://open.spotify.com/playlist/37i9dQZF1DX0Uv9tZ47pWo",]
```

```python
def big_audio_df(playlists_list):
    a_features_df = pd.DataFrame()
    
    for p in playlists_list:
        a_features_df = a_features_df.append(audio_features_df(p))

    return a_features_df
```

```python
# Create a dataframe of the consolidated audio features

audio_features_big_file = big_audio_df(playlists)
```
